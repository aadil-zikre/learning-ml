```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_activation(func):
    xs = np.arange(-5,5,0.2)
    ys = [func(i) for i in xs]
    plt.plot(xs, ys)
    plt.xticks(np.arange(-5,5,1))
    plt.yticks(np.arange(-5,5,1))
    plt.axhline(linewidth=2, color='black')
    plt.axvline(linewidth=2, color='black')
    plt.grid(alpha=0.5, markevery=1)
    plt.show()
```

```python
def activation_tanh(x):
    return np.tanh(x)
```

```python
plot_activation(activation_tanh)
```

```python
def activation_sigmoid(x):
    return 1/(1+np.exp(-1*x))
```

```python
plot_activation(activation_sigmoid)
```

```python
def activation_relu(x):
    return 0 if x <= 0 else x
```

```python
plot_activation(activation_relu)
```

```python
def activation_leaky_relu(x):
    leakage_coefficient = 0.1
    return leakage_coefficient * x if x <= 0 else x
```

```python
plot_activation(activation_leaky_relu)
```
